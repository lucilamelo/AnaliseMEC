1- Ler o arquivo.

```python
import pandas as pd
import seaborn as sns

arquivo = pd.read_csv("PDA_Dados_Cursos_Graduacao_Brasil.csv")
```

2- Resumo do arquivo.

```python
arquivo.info()
```

3- Exibir as linhas do arquivo. 

```python
arquivo.head()
```

4- Quantidade de cursos por codigo.


```python
n_total_cursos = arquivo['CODIGO_CURSO'].nunique()

print(f"Total de cursos: {n_total_cursos}")
```

5- Quantidade de cursos por Estado.

```python
cursos_estados = arquivo['UF'].value_counts()

print(cursos_estados)
```

6- Distribuição de cursos por Estado.

```python
import matplotlib.pyplot as plt

cursos_estados.plot(kind='bar')
plt.title('Distribuição de cursos por Estado')
plt.xlabel('Estado')
plt.ylabel('numero de Cursos')
plt.show()
```

7- Quantidade de cursos por modalidade.

```python

arquivo_sp = arquivo[arquivo['UF'] == 'SP']


cursos_por_categoria_sp = arquivo_sp['CATEGORIA_ADMINISTRATIVA'].value_counts().reset_index()
cursos_por_categoria_sp.columns = ['CATEGORIA_ADMINISTRATIVA', 'CODIGO_IES']

plt.figure(figsize=(10,6))
sns.barplot(x='CATEGORIA_ADMINISTRATIVA', y='CODIGO_IES', data=cursos_por_categoria_sp, palette='pastel')
plt.title('Número de Cursos por Categoria Administrativa da IES (São Paulo)')
plt.ylabel('Número de Cursos')
plt.xlabel('Categoria Administrativa')
plt.xticks(rotation=45)
plt.show()
```

```python

arquivo_mg = arquivo[arquivo['UF'] == 'MG']

cursos_por_categoria_mg = arquivo_mg['CATEGORIA_ADMINISTRATIVA'].value_counts().reset_index()
cursos_por_categoria_mg.columns = ['CATEGORIA_ADMINISTRATIVA', 'CODIGO_IES']

plt.figure(figsize=(10,6))
sns.barplot(x='CATEGORIA_ADMINISTRATIVA', y='CODIGO_IES', data=cursos_por_categoria_mg, palette='pastel')
plt.title('Número de Cursos por Categoria Administrativa da IES (Minas Gerais)')
plt.ylabel('Número de Cursos')
plt.xlabel('Categoria Administrativa')
plt.xticks(rotation=45)
plt.show()
```

```python
modalidade_cursos = arquivo['MODALIDADE'].value_counts()

print(modalidade_cursos)
```

8- Distribuição dos Cursos por Modalidade.

```python
modalidade_cursos.plot(kind='pie',autopct='%1.1f%%')
plt.title('Distribuição dos Cursos por Modalidade')
plt.show()
```

9- Top 10 cursos mais comuns.

```python
top_cursos = arquivo['NOME_CURSO'].value_counts().head(10)

print(top_cursos)

top_cursos.plot(kind= 'barh', color='skyblue')
plt.title('top 10 Cursos Mais Comuns')
plt.show()
```

10- Cursos mais pocurado em cada Estado.

```python
 
cursos_estado = arquivo.groupby(['UF', 'NOME_CURSO']).size().reset_index(name='contagem')


curso_mais_procurado = cursos_estado.loc[cursos_estado.groupby('UF')['contagem'].idxmax()]


print(curso_mais_procurado)


curso_mais_procurado.to_csv('cursos_mais_procurados_por_estado.csv', index=False)
```

11- Grafico dos cursos mais procurados. 

```python
import matplotlib.pyplot as plt

cursos_estado = arquivo.groupby(['UF', 'NOME_CURSO']).size().reset_index(name='contagem')

curso_mais_procurado = cursos_estado.loc[cursos_estado.groupby('UF')['contagem'].idxmax()]

plt.figure(figsize=(12, 8))
plt.barh(curso_mais_procurado['NOME_CURSO'] + " (" + curso_mais_procurado['UF'] + ")", curso_mais_procurado['contagem'], color='skyblue')
plt.title('Cursos Mais Procurados em Cada Estado')
plt.xlabel('Número de Cursos')
plt.ylabel('Curso (Estado)')
plt.tight_layout()
plt.show()
```

12- Numero de cursos por Categoria administrativa

```python

cursos_por_categoria = arquivo['CATEGORIA_ADMINISTRATIVA'].value_counts().reset_index()
cursos_por_categoria.columns = ['CATEGORIA_ADMINISTRATIVA', 'CODIGO_IES']


plt.figure(figsize=(10,6))
sns.barplot(x='CATEGORIA_ADMINISTRATIVA', y='CODIGO_IES', data=cursos_por_categoria, palette='pastel')
plt.title('Número de Cursos por Categoria Administrativa da IES')
plt.ylabel('CODIGO_IES')
plt.xlabel('CATEGORIA_ADMINISTRATIVA')
plt.xticks(rotation=45)
plt.show()
```

13- Top 10 Areas de conhecimento com mais cursos

```python

cursos_por_area = arquivo['AREA_OCDE'].value_counts().head(10).reset_index()
cursos_por_area.columns = ['AREA_OCDE', 'CODIGO_IES']

plt.figure(figsize=(10,6))
sns.barplot(y='AREA_OCDE', x='CODIGO_IES', data=cursos_por_area, palette='coolwarm')
plt.title('Top 10 Áreas de Conhecimento com Mais Cursos')
plt.xlabel('Número de Cursos')
plt.ylabel('Área OCDE')
```

14- Distribuição de vagas autorizadas por Região

```python

vagas_por_regiao = arquivo.groupby('REGIAO')['QT_VAGAS_AUTORIZADAS'].sum().reset_index()

plt.figure(figsize=(8,8))
plt.pie(vagas_por_regiao['QT_VAGAS_AUTORIZADAS'], labels=vagas_por_regiao['REGIAO'], autopct='%1.1f%%', startangle=140)
plt.title('Distribuição de Vagas Autorizadas por Região')
plt.axis('equal')
plt.show()
```
